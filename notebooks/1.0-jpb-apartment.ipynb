```python
import os
import sys
from pathlib import Path

PROJ_ROOT = os.path.join(os.pardir)
src_dir = os.path.join(PROJ_ROOT, 'src')
sys.path.append(src_dir)

from build_model import LoadResizeImages, HubStylizedImage
```

```python
CONTENT_PATH = Path('contents')
content_name = 'apartment1.jpg'
content_path = os.path.join(PROJ_ROOT, CONTENT_PATH, content_name)

STYLE_PATH = Path('styles')
style_name = 'krystal-ng.jpg'
style_path = os.path.join(PROJ_ROOT, STYLE_PATH, style_name)
```

```python
apartment = LoadResizeImages(content_path, style_path, 512)
```

```python
apartment.show_image(apartment.content_img)
```

```python
apartment.content_img = apartment.rotate_image(apartment.content_img, 3)
```

```python
apartment.show_image(apartment.content_img)
```

```python
apartment.show_image(apartment.style_img)
```

```python
hub = HubStylizedImage(apartment.content_img,
                       apartment.style_img,
                       'https://tfhub.dev/google/magenta/'
                       'arbitrary-image-stylization-v1-256/1')
```
